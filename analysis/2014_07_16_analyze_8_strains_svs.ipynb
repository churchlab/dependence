```python
import re

import pandas as pd
```

```python
all_sv_data_df = pd.read_csv('data/2014_07_16_lumpy_on_8_latest_strains.no_schema.vcf', delimiter='\t')
```

Rename column headers to sample names.

```python
uid_to_label_mapping_df = pd.read_csv('data/mapping_uid_to_label_for_8_latest_samples.csv', header=None)
uid_to_label_mapping_df = uid_to_label_mapping_df.set_index(0)
rename_map = {}
for key in all_sv_data_df.keys():
    if key in uid_to_label_mapping_df.index:
        rename_map[key] = str(uid_to_label_mapping_df.loc[key][1])
all_sv_data_df = all_sv_data_df.rename(columns=rename_map)
all_sv_data_df = all_sv_data_df.rename(columns={'#CHROM': 'CHROM'})
all_sv_data_df.keys()
```

```python
SAMPLE_COL_NAMES = [u'adk.d6_Esc1', u'adk.d6_Esc2', u'tyrS.d8_Esc1', u'adk.d6_DEP', u'tyrS.d8_DEP', u'tyrS.d8_Esc2', u'tyrS.d8_Esc3', u'adk.d6_Esc3']
all_sv_data_df[SAMPLE_COL_NAMES][:3]
```

Rank by quality

```python
all_sv_data_df = all_sv_data_df.sort('QUAL', ascending=False)
```

```python
def print_data(df, start=0, end=50):
    for idx, row in df[start:end].iterrows():
        print row[['POS', 'REF', 'ALT', 'QUAL']]
        print row['INFO']
        print row[SAMPLE_COL_NAMES]
        print '>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>'
```

```python
print_data(all_sv_data_df, end=3)
```

Manual inspection of the SV data doesn't show diddly.

Let's instead try sorting by number of reads supporting the SV. We get this by DP from the INFO string.

```python
all_sv_data_df['DP'] = all_sv_data_df.INFO.apply(lambda i: int(re.search('DP=([\d]+)', i).group(1)))
```

Extract other useful metadata for sorting.

```python
all_sv_data_df['END_CHR'] = all_sv_data_df.INFO.apply(lambda i: re.search('END_CHR=([\w]+)', i).group(1))
all_sv_data_df['END'] = all_sv_data_df.INFO.apply(lambda i: int(re.search('END=([\d]+)', i).group(1)))
all_sv_data_df['SIZE'] = all_sv_data_df.END - all_sv_data_df.POS
```

```python
all_sv_data_df[['DP', 'END_CHR']][:20]
```

```python
all_sv_data_df = all_sv_data_df.sort('DP', ascending=False)
all_sv_data_df[:20].to_csv('data/2014_07_16_lumpy_on_8_latest_strains.no_schema.sorted_DP.csv')
```

```python
CP006698_only = all_sv_data_df[all_sv_data_df.END_CHR == 'CP006698']
print_data(CP006698_only, end=20)
```
