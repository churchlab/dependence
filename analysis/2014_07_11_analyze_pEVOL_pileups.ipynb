```python
def create_pileup_df(source):
    return pd.read_csv(os.path.join(PILEUP_DATA_DIR, source),
            delimiter='\t')
    
adk_dep_pileup_df = create_pileup_df('adk_d6_dep_9cc50650.mpileup.summary')
adk_esc1_pileup_df = create_pileup_df('adk_d6_esc1_b1c44bc4.mpileup.summary')
adk_esc2_pileup_df = create_pileup_df('adk_d6_esc2_974f0007.mpileup.summary')
adk_esc3_pileup_df = create_pileup_df('adk_d6_esc3_620fb7b6.mpileup.summary')
tyrS_dep_pileup_df = create_pileup_df('tyrs_d8_dep_e5252b91.mpileup.summary')
tyrS_esc1_pileup_df = create_pileup_df('tyrs_d8_esc1_c1ab2674.mpileup.summary')
tyrS_esc2_pileup_df = create_pileup_df('tyrs_d8_esc2_1ddbe3bc.mpileup.summary')
tyrS_esc3_pileup_df = create_pileup_df('tyrs_d8_esc3_435b14f6.mpileup.summary')
```

We make the assumption that single base mutations are to blame for issues that we're seeing. Let's plot, for each position, the max mutations of another base type that we observe. Then we'll see how this compares to annotation of the plasmid.

First, there are some mutations that are already present in the dep strain pEVOL relative to the reference genome. Identify these.  Cross-validate the mutations using the two dep strains.

```python
def identify_mutation_positions(pileup_df):
    mutations = {}
    for idx, row in pileup_df.iterrows():
        ref = row['ref']
        counts = row[['A', 'T', 'C', 'G']]
        counts.sort(ascending=False)
        if not counts.index[0] == ref:
            mutations[row['pos']] = counts.index[0]
    return mutations

adk_dep_mutations = identify_mutation_positions(adk_dep_pileup_df)
tyrS_dep_mutations = identify_mutation_positions(tyrS_dep_pileup_df)
assert adk_dep_mutations == tyrS_dep_mutations
```

```python
adk_dep_mutations.keys()
```

```python
def get_second_most_counts(pileup_df):
   
    def _get_max_mutation_data(row):
        ref = row['ref']
        counts = row[['A', 'T', 'C', 'G']]
        counts.sort(ascending=False)
        if not counts.index[0] == ref:
            if row['pos'] in adk_dep_mutations:
                assert counts.index[0] == adk_dep_mutations[row['pos']]
            else:
                print 'New mutation', row
        second_most_count = counts[1]
        second_most_alt = counts.index[1]
        return pd.Series({
            'pos': row['pos'],
            'ref': row['ref'],
            'cov': row['cov'],
            'second_most_count': counts[1],
            'second_most_alt': counts.index[1]
        })
    max_mutation_df = pileup_df.apply(_get_max_mutation_data, axis=1)
    return max_mutation_df

adk_dep_pileup_second_most_counts_df = get_second_most_counts(adk_dep_pileup_df)
adk_esc1_pileup_second_most_counts_df = get_second_most_counts(adk_esc1_pileup_df)
adk_esc2_pileup_second_most_counts_df = get_second_most_counts(adk_esc2_pileup_df)
adk_esc3_pileup_second_most_counts_df = get_second_most_counts(adk_esc3_pileup_df)
tyrS_dep_pileup_second_most_counts_df = get_second_most_counts(tyrS_dep_pileup_df)
tyrS_esc1_pileup_second_most_counts_df = get_second_most_counts(tyrS_esc1_pileup_df)
tyrS_esc2_pileup_second_most_counts_df = get_second_most_counts(tyrS_esc2_pileup_df)
tyrS_esc3_pileup_second_most_counts_df = get_second_most_counts(tyrS_esc3_pileup_df)
```

```python
NUM = 20
def get_top_noisy_positions_set(df):
    return set(df.sort('second_most_count', ascending=False)[:NUM]['pos'])

adk_dep_top_noisy_set = get_top_noisy_positions_set(adk_dep_pileup_second_most_counts_df)
adk_esc1_top_noisy_set = get_top_noisy_positions_set(adk_esc1_pileup_second_most_counts_df)
adk_esc2_top_noisy_set = get_top_noisy_positions_set(adk_esc2_pileup_second_most_counts_df)
adk_esc3_top_noisy_set = get_top_noisy_positions_set(adk_esc3_pileup_second_most_counts_df)
```
